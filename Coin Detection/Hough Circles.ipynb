# Hough Circles

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('6.jpg',1)
img_org=img.copy()
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img,"gray")

```

```python
img = cv2.GaussianBlur(img,(21,21),cv2.BORDER_DEFAULT)
plt.imshow(img,"gray")
```

```python

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,0.9,120,param1=50,param2=30,minRadius=60,maxRadius=0)
          
print (circles)
```

```python
circles_round = np.uint16(np.around(circles))
print(circles_round)
print(circles_round.shape)
print("I have  found "+str(circles_round.shape[1])+" coins")
```

```python
for i in circles[0,:]:
    cv2.circle(img_org,(i[0],i[1]),i[2],(255,0,0),7)
    cv2.circle(img_org,(i[0],i[1]),2,(0,0,255),3)

plt.imshow(img_org)
```
